```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

img = cv.imread('okey.jpg',cv.IMREAD_UNCHANGED)

img_grey = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

cv2_imshow(img_grey)

canny = cv.Canny(img_grey, 200, 60, 60*3)

#kontur bulma
contours, hier = cv.findContours(canny,cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)

#kontur gster
cv.drawContours(img,contours,-1,(0,255,0),3)
cv2_imshow(img)

#Ka nesne bulundu
print('Bulunan nesne: ',len(contours))


cv.waitKey(0)
cv.destroyAllWindows()
```
