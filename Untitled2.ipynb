```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

```

```python
img = cv2.imread('8.jpg',0)
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(img,kernel,iterations=1)
dilation = cv2.dilate(img,kernel,iterations=1)


```

```python
plt.subplot(131),plt.imshow(img),plt.title('Orijinal Image')
plt.subplot(132),plt.imshow(erosion),plt.title('Erosion Image')
plt.subplot(133),plt.imshow(dilation),plt.title('Dilation Image')

```

```python
img2 = cv2.imread('8.jpg',0)
newimage = cv2.line(img2,(0,0),(511,511),(255,255,255),5)
openning = cv2.morphologyEx(newimage,cv2.MORPH_OPEN,kernel)
closing = cv2.morphologyEx(newimage,cv2.MORPH_CLOSE,kernel)
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
tophat = cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
blackhat = cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
plt.subplot(161),plt.imshow(newimage),plt.title('New Image')
plt.subplot(162),plt.imshow(openning),plt.title('Openning Image')
plt.subplot(163),plt.imshow(closing),plt.title('Closing Image')
plt.subplot(164),plt.imshow(gradient),plt.title('Gradient Image')
plt.subplot(165),plt.imshow(tophat),plt.title('Top Hat Image')
plt.subplot(166),plt.imshow(blackhat),plt.title('Black Hat Image')

```
