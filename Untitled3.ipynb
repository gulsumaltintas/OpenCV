```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread("mario2.jpg", 0)
plt.subplot(121)
plt.imshow(img)

template = cv2.imread("1.jpg", 0)
h, w = template.shape
res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)

threshold = 0.96
loc = np.where( res >= threshold)
print (loc)
print (loc[::-1])
print (zip(*loc[::-1]))
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), 255, 3)

tus = cv2.waitKey(0)

if tus == 27:
    cv2.destroyAllWindows()
elif tus == ord('s'):
    cv2.imwrite("yenimario.jpg", img)
    cv2.destroyAllWindows()


plt.subplot(122)
plt.imshow(img)
plt.show()
```
